```python
# Finalize dataset
df_use = pd.read_csv('../data/data_derived/derived_geocodes_pymc.csv', index_col=False)  

df_use.head(3)
```

```python
df_use.columns
```

```python
df_use['count_rows']=1
df_use['count']=1

df_use['status_n']= 0
#df_use['status_n'].mask(df_use['total_events'] == 0, 0, inplace=True)
df_use['status_n'].mask(df_use['total_events'] >= 1, 1, inplace=True)

df_use['status_c']='x'
df_use['status_c'].mask(df_use['total_events'] == 0, 'Did Not Attend', inplace=True)
df_use['status_c'].mask(df_use['total_events'] > 0, 'Attended', inplace=True)

df_use['pre_oh'] = df_use['pre_oh'].fillna(0)
df_use['session_1'] = df_use['session_1'].fillna(0)
df_use['session_2'] = df_use['session_2'].fillna(0)
df_use['session_3'] = df_use['session_3'].fillna(0)
df_use['post_oh'] = df_use['post_oh'].fillna(0)
df_use['submitted_pr'] = df_use['submitted_pr'].fillna(0)


# code for renaming columns
# df_locinfo2 = df_locinfo2.rename(columns={'province_state': 'state-province',
#                         'Timestamp': 'timestamp',
#                        })


```
